# importing the dataframes

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_BTC = pd.read_csv("Data/BTC_Historical.csv", index_col=0)
```

```python
df_DOGE = pd.read_csv("Data/DOGE_Historical.csv", index_col=0)
```

```python
df_XRP = pd.read_csv("Data/XRP_Historical.csv", index_col=0)
```

```python
df_ETH = pd.read_csv("Data/ETH_Historical.csv", index_col=0)
```

```python
df_UNI = pd.read_csv("Data/UNI_Historical.csv", index_col=0)
```

# Functions for Processing

```python
def Difference_between_open_and_close(df):
    df['Difference between opening and closing'] = df['Closing Price (USD)'] - df['Opening Price (USD)']
```

```python
def exporting_dataframe(df, filename):
    df.to_csv(f'{filename}.csv', index=True)
```

# Adding a Column that shows difference between opening and closing

```python
Difference_between_open_and_close(df_BTC)
```

```python
Difference_between_open_and_close(df_DOGE)
```

```python
Difference_between_open_and_close(df_XRP)
```

```python
Difference_between_open_and_close(df_ETH)
```

```python
Difference_between_open_and_close(df_UNI)
```

# Graphing the transformation of bitcoin over time

```python
df_BTC.index = pd.to_datetime(df_BTC.index)

plt.figure(figsize=(10, 6))
plt.plot(df_BTC.index, df_BTC['Highest Price (USD)'], color='blue', marker='o', linestyle='-')
plt.title('Transformation of BTC HIGHEST Price over Time')
plt.xlabel('Date')
plt.ylabel('Highest Price (USD)')
plt.grid(True)

plt.savefig('Transformation of BTC HIGHEST Price over Time.png')
```

```python
df_BTC.index = pd.to_datetime(df_BTC.index)

plt.figure(figsize=(10, 6))
plt.plot(df_BTC.index, df_BTC['Lowest Price (USD)'], color='red', marker='o', linestyle='-')
plt.title('Transformation of BTC LOWEST Price over Time')
plt.xlabel('Date')
plt.ylabel('Lowest Price (USD)')
plt.grid(True)

plt.savefig('Transformation of BTC LOWEST Price over Time.png')
```
